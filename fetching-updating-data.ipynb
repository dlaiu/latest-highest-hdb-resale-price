# This notebook will update the csv of HDB resale prices on the last day of the month

```python
import requests
```

```python
from datetime import datetime

# Get today's date
today_date = datetime.today()

# Get the year and month from the date
today_year = today_date.year
today_month = today_date.month

# Format the date as "YYYY-MM"
formatted_date = today_date.strftime("%Y-%m")

formatted_date
```

```python
import json
query = {'month':formatted_date}
query = json.dumps(query)
url = f"https://data.gov.sg/api/action/datastore_search?resource_id=d_8b84c4ee58e3cfc0ece0d773c8ca6abc&filters={query}&limit=1000&sort=month desc"
print(url)
```

```python
response = requests.get(url)
data = response.json()
data
```

```python
len(data['result']['records'])
```

```python
data['result']['records'][-1]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data['result']['records'])
df
```

```python
for month in df['month']:
    if month != formatted_date:
        print(month)
```

```python
df_old = pd.read_csv("resaleflatprices.csv")
df_old
```

```python
df_updated = pd.concat([df_old, df], ignore_index=True)
df_updated
```

```python
df_updated.to_csv('resaleflatprices_updated.csv', index=False)
```
