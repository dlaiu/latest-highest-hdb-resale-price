```python
import pandas as pd
```

```python
df = pd.read_csv('resaleflatprices.csv')
df
```

```python
block_max_price = df.groupby(['street_name','block','flat_type'])['resale_price'].max()
block_max_price
```

```python
df_block_info = block_max_price.to_frame()
df_block_info
```

```python
df_block_info = df_block_info.reset_index()
```

```python
df_block_info
```

```python
df_block_info['address'] = df_block_info['block'] + ' ' + df_block_info['street_name']
df_block_info
```

```python
df_block_info.drop(['street_name','block'], axis=1, inplace=True)
df_block_info
```

```python
formatted_numbers = []
for price in df_block_info['resale_price']:
    formatted_number = "${:,.0f}".format(price)

    formatted_numbers.append(formatted_number)

df_block_info['resale_price'] = formatted_numbers
df_block_info
```

```python
df_geo = pd.read_json('map/geo-data.json')
df_geo = pd.json_normalize(df_geo['features'])
df_geo
```

```python
len(df_geo)
```

```python
all_articles = []
```

```python
for place in df_geo['properties.place']:
    df_temp = df_block_info[df_block_info['address'] == place]
    # print(df_temp['address'])
    article = { 'Address': place }

    for row in df_temp.itertuples():
        article[row.flat_type] = row.resale_price

    all_articles.append(article)

all_articles
```

```python
len(all_articles)
```

```python
df_geo['properties.article'] = ''
```

```python
for index, row in df_geo.iterrows():
    temp_articles = all_articles[index]
    temp_str = ''
    for key, value in temp_articles.items():
        temp_str += f"{key}: {value}\n"
    # print(temp_str)
    df_geo.at[index, 'properties.article'] = temp_str
    # print(row['properties.article'])

df_geo
```

```python
quar_25 = df_geo['properties.max_price'].quantile(0.25)
quar_75 = df_geo['properties.max_price'].quantile(0.75)
quar_50 = df_geo['properties.max_price'].quantile(0.50)

print(quar_25)
print(quar_75)
print(quar_50)
```

```python
def ranking_price(row):
    # quar_25 = row['properties.max_price'].quantile(0.25)
    # quar_75 = row['properties.max_price'].quantile(0.75)
    # quar_50 = row['properties.max_price'].quantile(0.50)
    
    if row['properties.max_price'] > quar_75:
        return 1
    elif quar_50 < row['properties.max_price'] < quar_75:
        return 2
    elif quar_25 < row['properties.max_price'] < quar_50:
        return 3
    else:
        return 4
```

```python
df_geo['properties.price_rank'] = df_geo.apply(lambda row: ranking_price(row), axis = 1)
df_geo
```

```python
import json
```

```python
ok_json = json.loads(df_geo.to_json(orient='records'))
ok_json
```

```python
def process_to_geojson(file):
    geo_data = {"type": "FeatureCollection", "features":[]}
    for row in file:
        this_dict = {"type": "Feature", "properties":{}, "geometry": {}}
        for key, value in row.items():
            key_names = key.split('.')
            if key_names[0] == 'geometry':
                this_dict['geometry'][key_names[1]] = value
            if str(key_names[0]) == 'properties':
                this_dict['properties'][key_names[1]] = value
        geo_data['features'].append(this_dict)
    return geo_data
```

```python
geojson = process_to_geojson(ok_json)
geojson
```
